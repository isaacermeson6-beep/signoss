```python
import tkinter as tk
from tkinter import messagebox

# -------- FUNÇÕES -------- #

def descobrir_signo(dia, mes):

    if (dia >= 21 and mes == 3) or (dia <= 19 and mes == 4):
        return "Áries", "Fogo"
    elif (dia >= 20 and mes == 4) or (dia <= 20 and mes == 5):
        return "Touro", "Terra"
    elif (dia >= 21 and mes == 5) or (dia <= 20 and mes == 6):
        return "Gêmeos", "Ar"
    elif (dia >= 21 and mes == 6) or (dia <= 22 and mes == 7):
        return "Câncer", "Água"
    elif (dia >= 23 and mes == 7) or (dia <= 22 and mes == 8):
        return "Leão", "Fogo"
    elif (dia >= 23 and mes == 8) or (dia <= 22 and mes == 9):
        return "Virgem", "Terra"
    elif (dia >= 23 and mes == 9) or (dia <= 22 and mes == 10):
        return "Libra", "Ar"
    elif (dia >= 23 and mes == 10) or (dia <= 21 and mes == 11):
        return "Escorpião", "Água"
    elif (dia >= 22 and mes == 11) or (dia <= 21 and mes == 12):
        return "Sagitário", "Fogo"
    elif (dia >= 22 and mes == 12) or (dia <= 19 and mes == 1):
        return "Capricórnio", "Terra"
    elif (dia >= 20 and mes == 1) or (dia <= 18 and mes == 2):
        return "Aquário", "Ar"
    elif (dia >= 19 and mes == 2) or (dia <= 20 and mes == 3):
        return "Peixes", "Água"
    else:
        return None, None


def calcular_ascendente(hora):
    signos = [
        "Áries", "Touro", "Gêmeos", "Câncer",
        "Leão", "Virgem", "Libra", "Escorpião",
        "Sagitário", "Capricórnio", "Aquário", "Peixes"
    ]
    indice = (hora // 2) % 12
    return signos[indice]


def calcular():
    try:
        dia = int(entry_dia.get())
        mes = int(entry_mes.get())
        hora = int(entry_hora.get())

        if not (1 <= dia <= 31 and 1 <= mes <= 12 and 0 <= hora <= 23):
            raise ValueError

        signo, elemento = descobrir_signo(dia, mes)
        ascendente = calcular_ascendente(hora)

        if signo:
            resultado.config(
                text=f"🔮 Signo: {signo}\n"
                     f"🔥 Elemento: {elemento}\n"
                     f"⭐ Ascendente: {ascendente}"
            )
        else:
            messagebox.showerror("Erro", "Data inválida!")

    except ValueError:
        messagebox.showerror("Erro", "Digite apenas números válidos!")


# -------- INTERFACE -------- #

app = tk.Tk()
app.title("Descubra seu Signo")
app.geometry("300x300")
app.resizable(False, False)

tk.Label(app, text="Dia de nascimento").pack(pady=5)
entry_dia = tk.Entry(app)
entry_dia.pack()

tk.Label(app, text="Mês de nascimento").pack(pady=5)
entry_mes = tk.Entry(app)
entry_mes.pack()

tk.Label(app, text="Hora de nascimento (0-23)").pack(pady=5)
entry_hora = tk.Entry(app)
entry_hora.pack()

tk.Button(app, text="Calcular", command=calcular).pack(pady=15)

resultado = tk.Label(app, text="", font=("Arial", 12))
resultado.pack(pady=10)

app.mainloop()
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML

# -------- INTERFACE IPYWIDGETS -------- #

# Widgets de entrada
dia_widget = widgets.IntSlider(min=1, max=31, value=1, description='Dia:')
mes_widget = widgets.IntSlider(min=1, max=12, value=1, description='Mês:')
hora_widget = widgets.IntSlider(min=0, max=23, value=0, description='Hora (0-23):')

# Widget de saída
output_widget = widgets.Output()

def calcular_ipywidgets(dia, mes, hora):
    with output_widget: # Redireciona a saída para o widget
        output_widget.clear_output() # Limpa a saída anterior
        try:
            # As funções descobrir_signo e calcular_ascendente precisam ser definidas
            # ou a célula que as define (ea92cf40) deve ser executada primeiro.
            if not (1 <= dia <= 31 and 1 <= mes <= 12 and 0 <= hora <= 23):
                print(HTML("<p style='color:red;'>Erro: Digite apenas números válidos!</p>"))
                return

            signo, elemento = descobrir_signo(dia, mes)
            ascendente = calcular_ascendente(hora)

            if signo:
                result_html = (
                    f"<p style='font-size:16px;'>🔮 Signo: <b>{signo}</b></p>"
                    f"<p style='font-size:16px;'>🔥 Elemento: <b>{elemento}</b></p>"
                    f"<p style='font-size:16px;'>⭐ Ascendente: <b>{ascendente}</b></p>"
                )
                print(HTML(result_html))
            else:
                print(HTML("<p style='color:red;'>Erro: Data inválida!</p>"))

        except ValueError:
            print(HTML("<p style='color:red;'>Erro: Digite apenas números válidos!</p>"))


# Botão para calcular
calcular_button = widgets.Button(description="Calcular Signo e Ascendente")

def on_button_click(b):
    calcular_ipywidgets(dia_widget.value, mes_widget.value, hora_widget.value)

calcular_button.on_click(on_button_click)

# Exibir a interface
display(widgets.VBox([
    widgets.Label(value="Descubra seu Signo e Ascendente"),
    dia_widget,
    mes_widget,
    hora_widget,
    calcular_button,
    output_widget
]))
```
